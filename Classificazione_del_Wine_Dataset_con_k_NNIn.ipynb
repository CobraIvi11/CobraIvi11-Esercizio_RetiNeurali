**Classificazione del Wine Dataset con k-NNIn**

In questo progetto applichiamo l'algoritmo k-Nearest Neighbors per classificare tre diverse tipologie di vino basandoci sulle loro caratteristiche chimiche.

**Obiettivi:**
*   Caricare il dataset e normalizzare i dati.
*   Addestrare il modello con $k=5$.
*   Analizzare i risultati tramite la Matrice di Confusione.








```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine #importo del wine quality dataset
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

# Caricamento e preparazione dati
wine = load_wine()
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.2, random_state=42)

# Scaling (essenziale per k-NN)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Creazione Modello (come da slide del prof)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

print("Modello addestrato con successo!")
```

**Analisi Grafica dei Risultati**

Nella matrice sottostante, confrontiamo le classi reali con quelle predette dal modello. La diagonale indica le predizioni corrette.

```python
from sklearn.metrics import confusion_matrix

y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Purples', fmt='g',
            xticklabels=wine.target_names, yticklabels=wine.target_names)
plt.title('Matrice di Confusione: Reale vs Predetto')
plt.xlabel('Predetto')
plt.ylabel('Reale')
plt.show()
```
